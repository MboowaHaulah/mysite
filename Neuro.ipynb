TRAIN,TEST AND VALIDATE

```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Neuroscience/Neuro.csv', parse_dates=['X1'], low_memory=False)

# Let's say we want to split the data in 80:10:10 for train:valid:test dataset
train_size = 0.7
valid_size=0.15

train_index = int(len(df)*train_size)

# First we need to sort the dataset by the desired column 
df.sort_values(by = 'X5', ascending=True, inplace=True)

df_train = df[0:train_index]
df_rem = df[train_index:]

valid_index = int(len(df)*valid_size)

df_valid = df[train_index:train_index+valid_index]
df_test = df[train_index+valid_index:]

X_train, y_train = df_train.drop(columns='X5').copy(), df_train['X5'].copy()
X_valid, y_valid = df_valid.drop(columns='X5').copy(), df_valid['X5'].copy()
X_test, y_test = df_test.drop(columns='X5').copy(), df_test['X5'].copy()
        
print('Train data:',X_train.shape), print(y_train.shape)
print('Validation data:',X_valid.shape), print(y_valid.shape)
print('Test data:' ,X_test.shape), print(y_test.shape)
```

TUNE HYPERPARAMETERS

```python
import pandas as pd
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
  
# defining parameter range
param_grid = {'C': [0.1, 1, 10, 100, 1000], 
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf']} 
  
grid = GridSearchCV(SVC(), param_grid, refit = True, verbose = 3)
  
# fitting the model for grid search
```

```python
# print best parameter after tuning
print(grid.best_params_)
  
# print how our model looks after hyper-parameter tuning
print(grid.best_estimator_)
```
